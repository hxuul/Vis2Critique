```python
import pandas as pd
```

```python
data=pd.read_csv('map2.csv')
```

```python
data.head(20)
```

```python
data_pop = pd.read_csv('population data.csv')
```

```python
data_pop_clean = data_pop[['Country Name', '2023']]
data_pop_clean = data_pop_clean.rename(columns={'2023': 'population'})
```

```python
data_pop_clean.head(5)
```

```python
combined_data = pd.merge(data, data_pop_clean, left_on='Country of Birth', right_on='Country Name', how='left')
```

```python
combined_data
```

```python
combined_data = combined_data.drop(columns=['Country Name'])
```

```python
combined_data
```

```python
updated_population_data = {
    'Vietnam': 99660133,
    'Yugoslavia': 23000000,  # Approximate population for Yugoslavia
    'Iran': 90102637,
    'Turkey': 84980000,
    'Syria': 22130000
}

# Update the population column for countries where population is NaN
for country, population in updated_population_data.items():
    combined_data.loc[(combined_data['Country of Birth'] == country) & combined_data['population'].isna(), 'population'] = population

# Display the updated dataframe to verify changes
combined_data.head(20)
```

```python
combined_data.to_csv('updated_combined_data.csv', index=False)
```
